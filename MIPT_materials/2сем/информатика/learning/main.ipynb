```python
import matplotlib.pyplot as plt
import numpy as np

n_rows = 20     #  "" 
n_cols = 10     #     
X = np.linspace(0, 5, n_rows)   #   
Y = np.zeros((n_rows, n_cols))  #   
for i in range(n_cols):     #  
            Y[:,i] = np.sin(X) + 0.3 * np.random.randn(20)
Y_mean = Y.mean(axis=1)     #    
Y_std = Y.std(axis=1, ddof=1)   #  
plt.figure(figsize=(9, 6))  #   
plt.errorbar(X, Y_mean, yerr=Y_std, fmt='o-g',capsize=4, elinewidth=1, ecolor='red',label=' ')
plt.plot(X, np.sin(X), 'b', label = ' ')
plt.legend()
plt.axes([.18, .18, .35, .25])  #  
plt.plot(X,Y)   #   
plt.text(0, -1, ' ')
plt.show()
        
```

```python
import re 
pattern = r'((\d)(\d))((\d)(\d))' 
string = r'123456789' 
match = re.finditer(pattern, string) 
for i in match: 
    for j in range(0,7):
        print(f' {j} >{i[j]}<   {i.start(j)}  {i.end(j)}')
```
