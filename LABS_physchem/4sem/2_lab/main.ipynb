```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parser
df = open("./легенды_в_чате.txt")
a = []
for i in range(4, 18):
    a.append(df.readlines())
a = a[0]
b = [[] for i in range(3, 18)]
iter = 0
for i in a:
    if i == ' \n':
        iter+=1
    else:
        b[iter].append(i)
b = b[1:]
x = [[] for i in range(14)]
y = [[] for i in range(14)]
for i in range(14):
    b[i] = b[i][2:]
b[0]
for i in range(14):
    for j in range(350-350, 701-350):
        b[i][j] = b[i][j][:-1].split(' ')
        b[i][j][0] = float(b[i][j][0])
        b[i][j][1] = float(b[i][j][1])
        x[i].append(b[i][j][0])
        y[i].append(b[i][j][1])
y[0]
```

```python
labels = ['МО:2.0ml', 'МО:1.5ml', 'МО:1.0ml', 'МО:0.5ml']
plt.figure(figsize=(10,6))
plt.title('кислотные(HCl:5ml)')
plt.ylabel('optic_den')
plt.xlabel('wave_len, nm')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
it = 0
for i in [0, 4, 1, 3]:
    plt.errorbar(x[i], y[i], yerr=0, xerr=0, label = labels[it])
    it+=1
# p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
# xx=np.linspace(min(x), max(x), 100)
# yy=np.polyval(p, xx)
# plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig('acid.png')

aaa = round(sum([x[0][y[0].index(max(y[0]))], x[1][y[1].index(max(y[1]))], x[3][y[3].index(max(y[3]))], x[4][y[4].index(max(y[4]))]])/4)
plt.vlines(aaa, 0, 1.1)
plt.show()
aaa
```

```python
plt.figure(figsize=(10,6))
plt.title('щелочные(NaOH:5ml)')
plt.xlabel('концентрация, ')
plt.ylabel('D')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
x_, y_ = [2.5*0.0782, 2.0*0.0782, 1.5*0.0782, 1.0*0.0782], [y[5][118], y[6][118], y[7][118], y[8][118]]
# x_, y_ = [2.5*0.0782, 2.0*0.0782, 1.5*0.0782, 1.0*0.0782], [y[5][127], y[6][127], y[7][127], y[8][127]]
plt.errorbar(x_, y_, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x_, y_, 1)
xx=np.linspace(min(x_), max(x_), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig('D_щел.png')
plt.show()
print(x_)
print(y_)
```
