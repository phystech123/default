```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
c = [1, 1/2, 1/4, 1/8, 1/16][:4]
c = [i/100 for i in c]
l = [18.5, 26.0, 35.8, 51.1, 89.5][:4]
y = [1/i for i in l]
x = [c[i]*l[i] for i in range(len(c))]

plt.figure()
plt.ylabel(r"$1 / \lambda$")
plt.xlabel(r"$c \cdot \lambda$")
xx = np.linspace(min(x), max(x), 100)
p = np.polyfit(x, y, 1)
yy = np.polyval(p, xx)
plt.grid()
plt.plot(xx, yy, label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.scatter(x, y, color="red")
plt.legend()
plt.savefig("./image.png")
plt.show()
```
