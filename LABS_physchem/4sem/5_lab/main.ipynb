```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=[100, 200, 300, 350, 400, 450]
y1=[4.13, 3.96, 3.87, 3.84, 3.83, 3.82]
x2=[0, 100, 200, 250, 300, 350, 400, 450]
y2=[4.56, 4.30, 4.23, 4.21, 4.19, 4.17, 4.15, 4.15]
x3=[0, 100, 200, 300, 400]
y3=[4.65, 4.20, 4.15, 4.05, 4.04]
plt.figure()
plt.scatter(x1, y1, label="NaCl")
plt.scatter(x2, y2, label="LiCl")
plt.scatter(x3, y3, label="KCl")
plt.legend()
# plt.title("")
plt.xlabel("V")
plt.ylabel("pH")
plt.grid()
plt.show()
```
