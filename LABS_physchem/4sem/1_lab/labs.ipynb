```python
# some import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
import os
```

```python
df = pd.read_csv(
                "data/1.csv"
              )
y=df['T']
x=df['t']
plt.figure(figsize=(10,6))
plt.title('KCl, навеска 2г')
plt.ylabel('Температура, градусы Цельсия')
plt.xlabel('Время, секунды')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')

p=np.polyfit(x[:8], y[:8], 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')

p=np.polyfit(x[-5:], y[-5:], 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.savefig("./pictures/1")
plt.legend()
plt.show()
```
