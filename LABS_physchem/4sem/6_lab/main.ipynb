```python
# parser
n = 10
df = open("./data/with.txt")
a = []
a.append(df.readlines())
a=a[0]
b = [[] for i in range(n)]
iter = 0
for i in a:
    if i == ' \n':
        iter+=1
    else:
        b[iter].append(i)
# b = b[1:]
x = [[] for i in range(n)]
y = [[] for i in range(n)]
for i in range(n):
    b[i] = b[i][2:]
for i in range(n):
    for j in range(200-200, 401-200):
        b[i][j] = b[i][j][:-1].split(' ')
        b[i][j][0] = float(b[i][j][0])
        b[i][j][1] = float(b[i][j][1])
        x[i].append(b[i][j][0])
        y[i].append(b[i][j][1])
y[n-1]





C = np.array([0.44, 0.88, 1.32, 1.76, 2.2, 5.5][:])
C_ =  np.array([y[i][70]/0.7799 for i in [0,1,2,3,-2,-1]])
G = (C-C_)*0.1/0.025
```

```python
x_ = 1/C
y_ = 1/G
x_ = x_[:-1]
y_ = y_[:-1]
plt.figure(figsize=(10,6))
plt.title('')
plt.xlabel('1/C')
plt.ylabel('1/G')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
plt.errorbar(x_, y_, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x_, y_, 1)
xx=np.linspace(0, max(x_), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig('.png')
plt.show()
print(f'G_m={1/p[1]}')
print(f'K={p[1]/p[0]}')
print(pow(np.dot((x_-np.mean(x_)),(y_-np.mean(y_))),2)/(np.dot(x_-np.mean(x_),x_-np.mean(x_))*np.dot(y_-np.mean(y_), y_-np.mean(y_))))
```

```python
x_ = np.log(C)
y_ = np.log(G)
x_ = x_[:-1]
y_ = y_[:-1]
plt.figure(figsize=(10,6))
plt.title('')
plt.xlabel('ln(C)')
plt.ylabel('ln(G)')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
plt.errorbar(x_, y_, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x_, y_, 1)
xx=np.linspace(min(x_), max(x_), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig('.png')
plt.show()
print(f'K={np.exp(p[1])}')
print(f'n={p[0]}')
print(pow(np.dot((x_-np.mean(x_)),(y_-np.mean(y_))),2)/(np.dot(x_-np.mean(x_),x_-np.mean(x_))*np.dot(y_-np.mean(y_), y_-np.mean(y_))))
```
