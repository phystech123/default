```python
# some import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
import os
```

```python
# reading of data

df = pd.read_excel(
                "data/data.xlsx", 
                sheet_name="опыт1"
              )
# df.loc[len(df.index)] = df.mean(axis=0)
df = df.loc[:, ["I, mA", "U, Upp", "чаостота"]]
df["1/ksi*2"] = (df["чаостота"]*df["I, mA"]*0.001/df["U, Upp"])**2
df["nu*2"] = df["чаостота"]*df["чаостота"]
df
```

```python
# data analise

y=df['1/ksi*2'].tolist()
y = y[:3] + y[4:]
x=df['nu*2'].tolist()
x = x[:3] + x[4:]

plt.figure(figsize=(10,6))
# plt.title('Зависимость teta от R')
plt.xlabel('nu^2')
plt.ylabel('1/ksi^2')
plt.grid()
plt.errorbar(x, y, yerr=0.000004, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("./data/data1.png", bbox_inches='tight')
plt.show()
# gl1 = f'k={1/p[0]:.6f}'
# print(gl1)
# print(p)
print(sum([i**2 for i in y])/sum([i**2 for i in x]))
```

```python
# reading of data
import math

df = pd.read_excel(
                "data/data.xlsx", 
                sheet_name="опыт2"
              )

df["1/ksi*2"] = (df["nu"]*df["I"]*0.001/df["U"])**2
df["nu*2"] = df["nu"]**2
df["psi without PI"] = df["psi without PI"]*math.pi - math.pi/2
df
# df.loc[len(df.index)] = df.mean(axis=0)
df
```

```python
# data analise

y=df['psi without PI'].tolist()
y = [math.tan(i) for i in y[:-8]]
print(y)
x=df['nu'].tolist()[:-8]

plt.figure(figsize=(10,8))
# plt.title('Зависимость teta от R')
plt.xlabel('nu')
plt.ylabel('tg(psi)')
plt.grid()
plt.errorbar(x, y, yerr=0.000004, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("./data/data3.png")
plt.show()
# gl1 = f'k={1/p[0]:.6f}'
# print(gl1)
# print(p)
print(sum([i**2 for i in y])/sum([i**2 for i in x]))
```
