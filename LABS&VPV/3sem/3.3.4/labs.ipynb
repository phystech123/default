```python
# data analise

# y=df['nF'].tolist()
# x=df['T'].tolist()

# x=df['R'].tolist()
# y=df['teta'].tolist()

y=df['k'].tolist()
x=df['I'].tolist()
# x.pop()
# y.pop()
# x.pop()
# y.pop()
# x.pop()
# y.pop()
plt.figure(figsize=(10,6))
# plt.title('Зависимость teta от R')
plt.xlabel('I, мА')
plt.ylabel('k, мВ/мТс')
plt.grid()
plt.errorbar(x, y, yerr=0.000004, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
# gl1 = f'k={1/p[0]:.6f}'
# print(gl1)
# print(p)
print(sum([i**2 for i in y])/sum([i**2 for i in x]))
```
