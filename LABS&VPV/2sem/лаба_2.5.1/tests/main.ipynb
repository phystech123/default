```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('rd.csv')
df['t']+=273
df['dp']-=57
df['dp']=df['dp']*0.2*9.8
df['sigma'] = df['dp']*1.2/4000

df['x2'] = df['t'] ** 2
df['y2'] = df['sigma'] ** 2
df['xy'] = df['sigma'] * df['t']

df.loc[len(df.index)] = df.mean(axis=0)
# lis=df['u'].tolist()
# l=[round(float(i)/42.4, 5) for i in lis]
# df1=pd.DataFrame({'dt':l})
# df['dt']=df1
df
```

```python
x=df['t'].tolist()[1:-1]
y=df['sigma'].tolist()[1:-1]
plt.figure(figsize=(10,6))
plt.title(' sigma  t')
plt.xlabel('t')
plt.ylabel('sigma')
plt.grid()
plt.errorbar(x, y, yerr=0.0005, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
```

```python
x=df['t'].tolist()[1:-1]
y=(0.00017*df['t']).tolist()[1:-1]
plt.figure(figsize=(10,6))
plt.title('q  t')
plt.xlabel('t')
plt.ylabel('q')
plt.grid()
plt.errorbar(x, y, yerr=0.00, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
```
