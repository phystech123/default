```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('установка1/t20.csv')
n0 = df['f'].tolist()[0]
df['df'] = df['f'] - n0
df ['xy'] = df['df'] * df['n']
df['<x2>'] = df['n']**2




x=df['n'].tolist()
y=df['df'].tolist()
plt.figure(figsize=(10,6))
plt.title('Зависимость df от n')
plt.xlabel('n')
plt.ylabel('df')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0] + 1, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
gl1 = f'k={p[0]:.6f}'
print(gl1)
print(sum([(y[i] * x[i]) for i in range(len(y))])/len(x))
print(sum([y[i] for i in range(len(y))])/len(x))
print(sum([x[i] for i in range(len(x))])/len(x))
print(sum([(x[i])**2 for i in range(len(y))])/len(x))
print((sum([x[i] for i in range(len(y))])/len(x))**2)
df
```
