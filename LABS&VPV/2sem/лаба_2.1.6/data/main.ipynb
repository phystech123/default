```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('data25.csv')
# df['dp']=df['dp']-1
lis=df['u'].tolist()
l=[round(float(i)/40.7, 5) for i in lis]
df['dt']=pd.DataFrame({'dt':l})*1000
df['dp']*=-100000


df['x2'] = df['dp'] ** 2
df['y2'] = df['dt'] ** 2
df['xy'] = df['dt'] * df['dp']

df.loc[len(df.index)] = df.mean(axis=0)
df
```

```python
x=df['dp'].tolist()[:-1]
y=df['dt'].tolist()[:-1]
plt.figure(figsize=(10,6))
plt.title('температура 25 градусов')
plt.xlabel('dP')
plt.ylabel('dT, K')
plt.grid()
plt.errorbar(x, y, yerr=0.005/40.7, xerr=5000, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 500)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.11f}, b={p[1]:.4f}')
plt.legend()
plt.show()
print(f'k={p[0]:.11f}')
```
