```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('data1.csv')
df1=df.iloc[:41,:]
df1['lnp']=np.log(df1['p'].tolist())

# lis=df['u'].tolist()
# l=[round(float(i)/42.4, 5) for i in lis]
# df1=pd.DataFrame({'dt':l})
# df['dt']=df1



x=df1['t'].tolist()
y=df1['lnp'].tolist()
plt.figure(figsize=(10,6))
plt.title('Эксраполяция')
plt.xlabel('1/P, Па^-1')
plt.ylabel('D')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(0, max(x), 10000)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
a = 1/100191.76
b=p[0]*a+p[1]
print(b)
plt.scatter([a, a], [b, 0.00007], color='g')
plt.legend()
plt.show()
print(f'k={p[0]:.6f}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('231.xlsx')
df=df.iloc[2:44,3:5 ]
df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'lndp'}, inplace=True)
```

```python
x=df['t'].tolist()
xer=x[0:7]
x=x[7:]
y=df['lndp'].tolist()
yer=y[0:7]
y=y[7:]
plt.figure(figsize=(10,6))
plt.title('Ухудшение 1')
plt.xlabel('t, c')
plt.ylabel('P, мм.рт.ст * 10^-5')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 10000)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.2f}, b={p[1]:.2f}')
plt.scatter(x, y, color='g')
plt.scatter(xer, yer, color='r')
plt.legend()
plt.show()
```
