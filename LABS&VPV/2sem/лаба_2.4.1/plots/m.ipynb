```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('coord.csv')
df ['xy'] = df['lnP'] * df['1/T']
df['<x2>'] = df['1/T']**2
df['<y2>'] = df['lnP']**2
df.loc[ len(df.index )] = df.mean(axis=0)




x=df['1/T'].tolist()[:-1]
y=df['lnP'].tolist()[:-1]
plt.figure(figsize=(10,6))
plt.title('Зависимость lnP от  1/T')
plt.xlabel('1/T')
plt.ylabel('lnP')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0] + 1, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
gl1 = f'k={p[0]:.6f}'
print(sum([(y[i] * x[i]) for i in range(len(y))])/len(x))
print(sum([y[i] for i in range(len(y))])/len(x))
print(sum([x[i] for i in range(len(x))])/len(x))
print(sum([(x[i])**2 for i in range(len(y))])/len(x))
print((sum([x[i] for i in range(len(y))])/len(x))**2)
df
```
