```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('data1.csv')
df['N']=df['I'] * df['U']
df['x2'] = df['N'] ** 2
df['y2'] = df['dT'] ** 2
df['xy'] = df['dT'] * df['N']
df.loc[len(df.index)] = df.mean(axis=0)
# lis=df['u'].tolist()
# l=[round(float(i)/42.4, 5) for i in lis]
# df1=pd.DataFrame({'dt':l})
# df['dt']=df1
df


```

```python
x=df['N'].tolist()
y=df['dT'].tolist()
plt.figure(figsize=(10,6))
plt.title('Зависимость dT от N')
plt.xlabel('N')
plt.ylabel('dT')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
# p = [p[0]+0.01, p[1]]
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.show()
gl1 = f'k={1/p[0]:.6f}'
print(gl1)
print(sum([i**2 for i in y])/sum([i**2 for i in x]))
```
