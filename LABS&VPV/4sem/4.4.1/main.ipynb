```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
y1 = np.array([
    np.sin(np.deg2rad(11 + 37/60 + 30/3600)),
    np.sin(np.deg2rad(12 + 33/60 + 18/3600)),
    np.sin(np.deg2rad(14 + 12/60 + 00/3600)),
    np.sin(np.deg2rad(15 + 48/60 + 58/3600)),
    np.sin(np.deg2rad(16 + 44/60 + 28/3600)),
    np.sin(np.deg2rad(16 + 48/60 + 8/3600)),
    np.sin(np.deg2rad(18 + 8/60 + 18/3600)),
    np.sin(np.deg2rad(20 + 11/60 + 27/3600))
])

y2 = np.array([
    np.sin(np.deg2rad(168 + 19/60 + 14/3600)),
    np.sin(np.deg2rad(167 + 24/60 + 55/3600)),
    np.sin(np.deg2rad(165 + 46/60 + 30/3600)),
    np.sin(np.deg2rad(164 + 9/60 + 44/3600)),
    np.sin(np.deg2rad(163 + 14/60 + 34/3600)),
    np.sin(np.deg2rad(163 + 10/60 + 43/3600)),
    np.sin(np.deg2rad(161 + 51/60 +41/3600)),
    np.sin(np.deg2rad(159 + 49/60 + 43/3600))
])

l = np.array([
    404.7,
    435.8,
    491.6,
    546.1,
    577.0,
    579.1,
    623.4,
    690.7
])
# l*=1e-9

plt.scatter(l, y1)
plt.plot(l, y1)
```

```python
y = l
x = (y1+y2)/2
plt.title(r'$\lambda$ от $\sin{\varphi_m}$')
plt.xlabel(r'$\sin{\varphi_m}$')
plt.ylabel(r'$\lambda$, мкм')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("data1.png")
plt.show()
```

```python
dx = pd.DataFrame([l, y1, y2])
dx = dx.T
print(dx.to_latex())
```

```python
m = np.array([1,2,3,4])
ph1 = np.array([163 + 10/60 + 43/3600, 215 + 18/60 + 14/3600, 240 + 16/60 + 38/3600, 120 + 2/60 + 30/3600])
ph2 = np.array([163 + 14/60 + 34/3600, 215 + 27/60 + 18/3600, 240 + 35/60 + 20/3600, 120 + 23/60 + 19/3600])
ph = np.deg2rad(np.abs(ph1-ph2))/(2.1*1e-7)

x = m[:3]
y = ph[:3]

plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("data2.png")
plt.show()
```

```python
dx = pd.DataFrame([m, ph1, ph2]).T
print(dx.to_latex())
```
