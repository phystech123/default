```python
x = [1115, 1232, 1280, 1325, 1365, 1405][1:]
x = [i-1115 for i in x]
y = [64, 61, 53, 46, 38, 33][1:]
y = [np.log(i) for i in y]


plt.figure(figsize=(10,6))
plt.title('показания микроамперметра от размера щели')
plt.ylabel('Ток, мкА')
plt.xlabel(r'расстояние, мкм$*10^{-1}$')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()

plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
plt.savefig("data/data2.png")
plt.show()
```

```python
x = [980, 1040, 1085, 1170, 1315, 1437, 1511]
x = [(i-980+1465-1115)/100 for i in x]
y = [33, 26, 20, 15, 10, 5, 3]
y = [np.log(i/33) for i in y]

plt.figure(figsize=(10,6))
plt.title('показания микроамперметра от размера щели')
plt.ylabel('Ток, мкА')
plt.xlabel(r'расстояние, мкм')
plt.grid()
plt.grid(which='minor')
plt.minorticks_on()
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()

plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
plt.savefig("data/data3.png")
plt.show()
```
