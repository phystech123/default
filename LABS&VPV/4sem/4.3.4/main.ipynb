```python
# some import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
import os
```

```python
x = [50, 100, 150, 200, 300, 400, 500]
y = [2000, 4000, 6000, 8000, 12000, 16000, 20000]
plt.figure(figsize=(10,6))
plt.title('размер щели от размера изобрадения')
plt.ylabel('мкм')
plt.xlabel('мкм')
plt.grid()
plt.errorbar(x, y, yerr=0, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("data1.png")
plt.show()
```
