```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [2.1 ,2.9, 3.3, 4.2, 4.7, 5.2, 5.6]
y = [i*i for i in y]

plt.figure(figsize=(10,6))
plt.title('Зависимость радиуса кольца от номера')
plt.xlabel('N')
plt.ylabel('R, cm')
plt.grid()
plt.errorbar(x, y, yerr=2, xerr=0, fmt='ob')
p=np.polyfit(x, y, 1)
xx=np.linspace(min(x), max(x), 100)
yy=np.polyval(p, xx)
plt.plot(xx, yy, '-r', label=f'k={p[0]:.6f}, b={p[1]:.6f}')
plt.legend()
plt.savefig("./pictures/data1.png", bbox_inches='tight')
plt.show()
```
