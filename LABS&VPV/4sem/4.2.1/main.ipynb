```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
y1 = [7.93, 7.86, 7.69, 7.59, 7.48, 7.35, 7.18, 7.03]
y1 = [pow((i-a)/10, 2) for i in y1]
y2 = [8.09, 7.91, 7.80, 7.67, 7.54, 7.41, 7.25, 7.09]
y2 = [pow((i-a)/10, 2) for i in y2]
tt = 5
x = [j for j in range(tt+8, tt, -1)]
print(x)
a = 4.89
# b = 4.46

p1=np.polyfit(x, y1, 1)
xx1=np.linspace(min(x), max(x), 100)
yy1=np.polyval(p1, xx1)
plt.plot(xx1, yy1, '-b', label=f'k={p1[0]:.6f}, b={p1[1]:.6f}')


p2=np.polyfit(x, y2, 1)
xx2=np.linspace(min(x), max(x), 100)
yy2=np.polyval(p2, xx2)
plt.plot(xx2, yy2, '-r', label=f'k={p2[0]:.6f}, b={p2[1]:.6f}')

sns.scatterplot(x=x, y=y1)
sns.scatterplot(x=x, y=y2)


plt.title('Графики координат светлых и темных полос\n(синий - светлый)(красный - темный)')
plt.xlabel('номер полосы')
plt.ylabel(r'$mm^2, \text{расстояние}^2$')
# plt.xlim(min(x), max(x))
# plt.ylim(min(y), max(y))
plt.savefig("./oleg1")
```
