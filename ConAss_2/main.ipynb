```python
import json
with open('configs/1.json', 'r') as json_file:
    data = json.load(json_file)


phi = pd.read_csv("data/data_phi.csv", sep=";")[" x"][::]
phi_ = pd.read_csv("data/data_phi_.csv", sep=";")[" x"][::]

theta = pd.read_csv("data/data_theta.csv", sep=";")[" x"][::]
theta_ = pd.read_csv("data/data_theta_.csv", sep=";")[" x"][::]


g = data["params_sys"]["g"]
l = data["params_sys"]["l"]
m = data["params_sys"]["m"]

alpha_1 = data["params_sys"]["alpha_1"]
alpha_2 = data["params_sys"]["alpha_2"]

x_0 = data["params_sys"]["x_0"]
y_0 = data["params_sys"]["y_0"]
z_0 = data["params_sys"]["z_0"]


E = 1/2*m*(l**2)*((theta_**2)*(np.sin(phi)**2) + phi_**2) - m*g*l*np.cos(phi) + alpha_1*alpha_2/np.sqrt(l**2 + x_0**2 + y_0**2 + z_0**2 - 2*l*(x_0*np.sin(phi)*np.cos(theta) + y_0*np.sin(phi)*np.sin(theta) + z_0*np.cos(phi)))
plt.figure()
plt.title("Energy(time)")
plt.grid()
plt.ylabel('Energy')
plt.xlabel('time')
sns.scatterplot(x=pd.read_csv("data/data_phi.csv", sep=";")["t"][::], y=E)

# data = E[::10].to_latex(index=False)
# print(data)
plt.savefig("./output/energy(t).png")
```

```python
import json
import os
import tempfile
import subprocess
from concurrent.futures import ThreadPoolExecutor


BASE_CONFIG = {
    "params_model": {
        "algo": "RK4_for_biff",
        "h": 0.001,
        "rep": 200000,
        "skip": 100000,
        "files": {
            "phi": "data_for_biff/data_phi.csv",
            "phi_": "data_for_biff/data_phi_.csv",
            "theta": "data_for_biff/data_theta.csv",
            "theta_": "data_for_biff/data_theta_.csv"
        }
    },
    "params_sys": {
        "g": 9.81,
        "l": 1,
        "m": 1,
        "betta": 0.5,

        "alpha_1": 0.1,
        "alpha_2": 1,
        
        "x_0": 0.0,
        "y_0": 0.0,
        "z_0": 1.1
    },
    "init_cond": {
        "phi": 1.57,
        "phi_": 0,
        "theta": 0,
        "theta_": 5
    }
}




CPP_PROGRAM = "./build/main"
THREADS = 12 

main_list = []


def generate_config(alpha, theta_):
    # config = BASE_CONFIG.copy()
    config = json.loads(json.dumps(BASE_CONFIG))
    config["params_sys"]["alpha_1"] = alpha
    config["init_cond"]["theta_"] = theta_

    with tempfile.NamedTemporaryFile(mode='w', suffix='.json', delete=False) as f:
        json.dump(config, f, indent=4)
        return f.name


def run_cpp_program(config_path):
    """Запускает C++ программу с указанным конфигом"""
    result = subprocess.run(
        [CPP_PROGRAM, config_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True,
        check=True  # Автоматически генерирует исключение при ненулевом коде возврата
    )
    res = list(map(float, result.stdout.split()))
    with open(config_path, 'r') as json_file:
        cc = json.load(json_file)
        main_list.append((res[1], res[3], cc["params_sys"]["alpha_1"]))
    try:
        os.unlink(config_path)
        print(f"Временный файл {config_path} удален")
    except OSError as e:
        print(f"Ошибка при удалении {config_path}: {e}")

    return True

def main():
    alphas = [round(x * 0.1, 1) for x in range(-10, 11)]
    thetas_ = [x for x in range(0,100)]
    configs = [generate_config(alpha, theta_) for theta_ in thetas_ for alpha in alphas]
    
    with ThreadPoolExecutor(max_workers=THREADS) as executor:
        results = list(executor.map(run_cpp_program, configs))
    # for conf in configs:
    #     run_cpp_program(conf[0], conf[1])
    
    print(f"Завершено! Всего запусков: {len(results)}, успешных: {sum(1 for r in results if r == 0)}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.widgets import Slider, Button

# Исходные данные
data = main_list

# Фильтрация данных (удаление NaN значений)
filtered_data = [point for point in data if not (np.isnan(point[0]) or np.isnan(point[1]) or np.isnan(point[2]))]

# Разделение на координаты
x = [point[0] for point in filtered_data]
y = [point[1] for point in filtered_data]
z = [point[2] for point in filtered_data]

# Создание 3D графика
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Настройка визуализации
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=40, alpha=0.8, depthshade=True)
ax.set_xlabel('X Coordinate', fontsize=12, labelpad=10)
ax.set_ylabel('Y Coordinate', fontsize=12, labelpad=10)
ax.set_zlabel('Z Coordinate', fontsize=12, labelpad=10)
ax.set_title('Интерактивный 3D график', fontsize=14, pad=20)

# Добавление цветовой шкалы
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Значение Z', fontsize=10)

# Настройка вида по умолчанию
ax.view_init(elev=30, azim=45)

# Добавление пояснения
plt.figtext(0.5, 0.01, "Используйте левую кнопку мыши для вращения, правую для панорамирования, колесо для масштабирования", 
            ha="center", fontsize=10, bbox={"facecolor":"orange", "alpha":0.2, "pad":5})

# Настройка интерфейса
plt.subplots_adjust(bottom=0.2)

# Добавление элементов управления
ax_elev = plt.axes([0.25, 0.1, 0.65, 0.03])
ax_azim = plt.axes([0.25, 0.05, 0.65, 0.03])

slider_elev = Slider(ax_elev, 'Верт. угол', 0, 90, valinit=30, valstep=1)
slider_azim = Slider(ax_azim, 'Гориз. угол', 0, 360, valinit=45, valstep=1)

def update(val):
    ax.view_init(elev=slider_elev.val, azim=slider_azim.val)
    fig.canvas.draw_idle()

slider_elev.on_changed(update)
slider_azim.on_changed(update)

# Кнопка сброса
resetax = plt.axes([0.8, 0.15, 0.1, 0.04])
button = Button(resetax, 'Сброс', color='lightgoldenrodyellow', hovercolor='0.975')

def reset(event):
    slider_elev.reset()
    slider_azim.reset()
button.on_clicked(reset)

plt.show()
```
